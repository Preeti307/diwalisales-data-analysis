```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding="latin1")
df.head()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(["Status","unnamed1"],axis=1)
df.shape
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df["Amount"].dtype
```

```python
df["Amount"] = df["Amount"].astype("int")
```

```python
df["Amount"].dtype
```

```python
df.columns
```

```python
df.rename(columns={"Marital_Status":"Shadi"})
```

```python
ax = sns.countplot(data=df,x="Gender")
```

```python
ax = sns.countplot(data=df,x="Gender",hue="Gender")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_gen = df.groupby("Gender", as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

sns.barplot(data=sales_gen, x="Gender", y="Amount",hue="Gender")
```

```python
df.columns
```

```python
ax = sns.countplot(data=df, x="Age Group", hue="Gender")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
plt.figure(figsize=(12, 6), constrained_layout=True)
sales_gen = df.groupby("Age Group",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
ax = sns.barplot(data=sales_gen,x="Age Group",y="Amount" , palette="Set2")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.columns
```

```python
plt.figure(figsize=(20,7),constrained_layout=True)
sales_state = df.groupby("State", as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False)
plt.grid(ls='-.')

ax = sns.barplot(data=sales_state, x="State", y="Orders")

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
plt.figure(figsize=(20,7),constrained_layout=True)
sales_state = df.groupby("State", as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
# plt.grid(ls='-.')

ax = sns.barplot(data=sales_state, x="State", y="Amount")
```

```python
ax = sns.countplot(data=df, x="Marital_Status",hue="Gender")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# plt.figure(figsize=(20,7),constrained_layout=True)
sales_state = df.groupby("Marital_Status", as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False)
# plt.grid(ls='-.')

ax = sns.barplot(data=sales_state, x="Marital_Status", y="Orders")

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_state = df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(data=sales_state, x="Marital_Status",y="Amount",hue="Gender")
```
